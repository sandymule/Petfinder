```python
imgurl = []
source = []
city = []
```

```python
output=pd.DataFrame()

for pgnum in range(1, 7000):

    url='http://www.petfinder.com/pet-search?query_id=21789902-B35C-11E6-B1D0-C63111471E21&page_number=' + str(pgnum)

# get browser driver object
    browser=webdriver.Chrome()
    browser.get(url=url)
    html = browser.page_source
    html = html.replace("&lt;", "<").replace("&gt;", ">")
    soup = BeautifulSoup(html)
    browser.quit()
    
# dropdown menu
#browser.find_element_by_xpath("//select[@name='animal']/option[text()='Dog']").click()

    pginfo = soup.find('ul',{"id":"search-results", "class":"adoptablePets"}).find_all('div',{'class':'adoptablePets-item'})
    for item in pginfo:
        imgurl.append(item.find('figure').find('img')['src'])
        info = item.find('p',{"class":"rescue-info"})
        source.append(info.find('a').text)
        city.append(info.find('span').text)
```

```python
img_url_df = pd.DataFrame(imgurl)
source_df = pd.DataFrame(source)
city_df = pd.DataFrame(city)
```
